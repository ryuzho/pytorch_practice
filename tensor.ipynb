```python
print(tensor.add(5))
print(tensor)
print(tensor.add_(5))
print(tensor)
# memory 일부 절약 but 기록(history)가 즉시 삭제되어 derivative 계산에 문제 발생 가능, 사용을 권장하지 않음
```

```python
t = torch.ones(5)
print(f"t : {t}")
n = t.numpy()
print(f"n : {n}")
```

```python
t.add_(2)
print(t)
print(n)
# CPU상의 tensor와 numpy배열은 메모리 공간을 공유
```

```python
n = np.ones(5)
t = torch.from_numpy(n)
np.add(n,1,out=n)
print(n)
print(t)
```
